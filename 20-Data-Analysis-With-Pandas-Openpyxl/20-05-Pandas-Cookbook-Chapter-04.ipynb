# A.)  Merge on Left - Almost identical vlookup() in Excel 

# On_merged columns gotta share same data types

```python
df1.rename(index=str, columns={"Product Code":"EAN"}, inplace=True)
df1.columns
```

```python
df3 = pd.merge(df1, df2, on="EAN", how="left")
```

```python
df3
```

```python
print(df1.dtypes)
print(df2.dtypes)
#So see: two dataframes have the same data types on the column "EAN" - :float64
```

```python
df4 = pd.merge(df1, df2, on="EAN", how="right")
```

```python
df4.head(3)
```

```python
df5 = pd.merge(df1, df2, on="EAN", how="inner")
df5.shape
```
